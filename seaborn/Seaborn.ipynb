#Importing ibraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Getting sesborn datasets for visualisation**

```python
tips = sns.load_dataset('tips') #this will be imported as pandas dataframe
type(tips)
```

```python
tips.head()
```

#Visualizing the data

*   data -> which dataset to plot
*   x -> x axis
*   y -> y axis
* col -> on which col to seperate plot. here we will have two plots for two different time -> lunch and dinner
* hue -> for different colors
* style -> for different data pointers; like dot, corss, etc
* size -> for size of data pointers

```python
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

#Setting theme

#Count Plot

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(data=titanic, x='class')
```

#Bar charts

```python
sns.barplot(data=titanic, y='survived', x='sex', hue='class')
```

**This black lines represents confidence interval**

```python
# Default barplot with confidence intervals
sns.barplot(data=tips, x="day", y="total_bill")
plt.title("Barplot with Confidence Intervals")
plt.show()
```

```python
# Barplot without error bars
sns.barplot(data=tips, x="day", y="total_bill", ci=None)
plt.title("Barplot without Error Bars")
plt.show()
```

#Distribution Plot

1. Distribution Plot

```python
sns.displot(data=titanic, x='age')
```

2. Hist Plot

```python
sns.histplot(data=titanic, x='age', kde=True)
```

3. KDE Plot

```python
sns.kdeplot(titanic['age'])
plt.title("KDE Plot")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

#Correlation using HeatMap

```python
from sklearn import datasets
diabetes_dataset = datasets.load_diabetes()
db_df = pd.DataFrame(diabetes_dataset.data, columns=diabetes_dataset.feature_names)
```

```python
db_df.head()
```

```python
correlation = db_df.corr()
```



* cbar -> for color bar
* square -> for each cell
* fmt = '.1f' -> for formatting values, here one decimal point
* annot -> for annotations
* annot_kws -> for annotation values
* cmpa -> for color of map



```python
sns.heatmap(correlation, cbar=False, square=False, fmt='.1f', annot=True, annot_kws={'size': 20}, cmap='magma')
```

```python
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```
