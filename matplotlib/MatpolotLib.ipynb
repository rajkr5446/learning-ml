#Importing the library and basic plots

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#importing numpy for data
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
print(x)
print(y)
print(z)
```

```python
#plotting the data
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

#Adding titles and labels to the plot

```python
plt.plot(x, y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine wave')
plt.show()
```

```python
plt.plot(x, z)
plt.xlabel('Angle')
plt.ylabel('Cosine Value')
plt.title('Cosine wave')
plt.show()
```

#Plotting points
By default it plots line.

```python
#generating parabola data
x = np.linspace(-10, 10, 20)
y = x ** 2
print(x)
print(y)
```

**'r+'**
* -> the first character represents color here r means red.
* -> the second character represents symbol here +**bold text**

```python
plt.plot(x, y, 'r+')
plt.show()
```

#Plotting multiple lines in a plot

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

#Plotting Bars

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1]) # 0, 0 -> coordinates of origin of plot,   1, 1 -> x-axis and y-axis lenght
languages = ['English', 'Spanish', 'Latin', 'French', 'German']
people = [100, 80, 20, 120, 50]
ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NO OF PEOPLE")
plt.show()
```

#Plotting Piecharts

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

#Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='r')
ax.scatter(x, z, color='g')
plt.show()
```
