#Manipulating a DataFrame

1. Adding a column to a DataFrame

```python
breast_cancer_df['Has Cancer'] = breast_cancer.target
breast_cancer_df.head()
#notice how 'Has Cancer' column has been added
```

2. Removing a row -> **index=0 means 1st row and axis=0 for row**

```python
breast_cancer_df2 = breast_cancer_df.drop(index=0, axis=0)
breast_cancer_df2.head()
```

```python
#notice that df2 has one less row
print(breast_cancer_df.shape)
print(breast_cancer_df2.shape)
```

3. Removing a column -> **axis=1 for columns**

```python
breast_cancer_df3 = breast_cancer_df.drop(columns='Has Cancer', axis=1)
breast_cancer_df3.head()
```

```python
#notice that df3 has one less column
print(breast_cancer_df.shape)
print(breast_cancer_df3.shape)
```

4. Locating a row using index value
---
*   .iloc[0] -> first row
*   .iloc[1] -> second row
*   .iloc[-1] -> last row

```python
breast_cancer_df.iloc[0]
```

5. Locating a column


---


*   .iloc[:,0] -> first column
*   .iloc[:,1] -> second column
*   .iloc[:,-1] -> last column

```python
breast_cancer_df.iloc[:,-1]
```
